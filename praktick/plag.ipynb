```python
import filecmp
import os

def compare_files(file1, file2):
    # Функция для сравнения двух файлов
    return filecmp.cmp(file1, file2)

def remove_files_and_annotations(file_paths):
    for path in file_paths:
        try:
            os.remove(path)
            print(f"Удален файл: {path}")
        except OSError as e:
            print(f"Ошибка при удалении файла {path}: {e}")

def check_folder_for_duplicates(folder_path):
    # Функция для проверки папки на дубликаты файлов
    txt_files = [f for f in os.listdir(folder_path) if f.endswith('.txt')]

    i = 0
    while i < len(txt_files):
        j = i + 1
        while j < len(txt_files):
            txt_file1_path = os.path.join(folder_path, txt_files[i])
            txt_file2_path = os.path.join(folder_path, txt_files[j])

            if compare_files(txt_file1_path, txt_file2_path):
                # Файлы идентичны. Удаляем и .txt, и .ann файлы
                ann_file1_path = os.path.splitext(txt_file1_path)[0] + ".ann"
                ann_file2_path = os.path.splitext(txt_file2_path)[0] + ".ann"

                remove_files_and_annotations([txt_file2_path, ann_file2_path])
                
                # Убираем удаленные файлы из копии списка
                txt_files.pop(j)
            else:
                j += 1
        
        i += 1

# Замените 'путь_к_папке' на путь к вашей папке
folder_path = r"C:\Users\fobus\Desktop\dataSet"
check_folder_for_duplicates(folder_path)



```
